```python
import pandas as pd
import os
import numpy as np
```

```python
file = "HSK 1.tsv"
hanzi_df = pd.read_csv(
    os.path.join("./data/Old_HSK_tsvs/", file), delimiter="\t", header=None, names=["Traditional", "Simplified", "Pinyin", "Definition"]
)
unique_hanzi = []
unique_hanzi_df = pd.DataFrame(columns=["Simplified", "Traditional"])

hanzi_df.head(5)
for index, row in hanzi_df.iterrows():
    for simplified, traditional in zip(row["Simplified"], row["Traditional"]):
        #print(f"simplified: {simplified}")
        #print(f"traditional: {traditional}")
        if simplified not in unique_hanzi:
            unique_hanzi.append(simplified)
            unique_hanzi_df.loc[len(unique_hanzi_df)] = [row["Simplified"], row["Traditional"]]
```
